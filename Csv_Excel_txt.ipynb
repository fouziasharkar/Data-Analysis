## Opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

## Convertors

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('/content/IPL_Matches_2008_2022.csv',converters={'team1':rename})
```

## na_values parameter

## Loading a huge dataset in chunks

```python
dfs_data = pd.read_csv('aug_train.csv')
dfs_data.shape
```

```python
#this devides a big dataset into little chunks

dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
dfs_data.head()
```

```python
# by using for loop we can work on each chunk simulteneusly

for chunks in dfs:
  print(chunks.shape)
```
